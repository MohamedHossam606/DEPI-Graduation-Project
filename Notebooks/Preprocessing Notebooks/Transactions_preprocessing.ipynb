```python
from sklearn.preprocessing import LabelEncoder

# Custom LabelEncoder that skips NaN values
class LabelEncoderWithNaNs(LabelEncoder):
    def fit_transform(self, X):
        X_non_null = X[~pd.isnull(X)]
        labels = super().fit_transform(X_non_null)
        
        # Create an array to hold encoded values
        encoded = np.full_like(X, fill_value=np.nan, dtype=float)
        encoded[~pd.isnull(X)] = labels
        return encoded

# Instantiate the custom LabelEncoder
le = LabelEncoderWithNaNs()
```

```python
Transactions_df_imputed = Transactions_df.copy()
Transactions_df_imputed.head(1)
```

```python
Transactions_df_imputed['encoded_product_first_sold_date'] = le.fit_transform(Transactions_df_imputed['product_first_sold_date'])
Transactions_df_imputed.head(1)
```

```python
Transactions_df_imputed.loc[Transactions_df_imputed['product_first_sold_date'].isna()].head(1)
```

```python
Transactions_df_imputed.drop('product_first_sold_date', axis=1 ,inplace=True)
Transactions_df_imputed.isnull().sum()
```

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import OneHotEncoder

temp_df = Transactions_df_imputed.drop(['transaction_id', 'transaction_date'], axis=1)

numerical_cols = temp_df.select_dtypes(include='number').columns
categorical_cols = temp_df.select_dtypes(include='object').columns

onehot_encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_categorical = pd.DataFrame(onehot_encoder.fit_transform(temp_df[categorical_cols]),
                                   columns=onehot_encoder.get_feature_names_out(categorical_cols),
                                   index=temp_df.index)  # Make sure indices match the original dataframe

# Impute missing values in numerical columns
imputer = KNNImputer(n_neighbors=3)
imputed_numerical = pd.DataFrame(imputer.fit_transform(temp_df[numerical_cols]), 
                                 columns=numerical_cols, 
                                 index=temp_df.index)  # Ensure indices match

# Concatenate the imputed numerical columns and the encoded categorical columns
df_imputed = pd.concat([imputed_numerical, encoded_categorical], axis=1)

# Update the original dataframe with imputed values
Transactions_df_imputed[numerical_cols] = df_imputed[numerical_cols]

Transactions_df_imputed.head()
```

```python
# Inverse transform to get the original values back
Transactions_df_imputed['product_first_sold_date'] = le.inverse_transform(Transactions_df_imputed['encoded_product_first_sold_date'].round().astype(int))
Transactions_df_imputed.drop('encoded_product_first_sold_date', axis=1 ,inplace=True)
```

```python
Transactions_df_imputed['product_first_sold_date'] = pd.to_datetime(Transactions_df_imputed['product_first_sold_date'])
```

```python
Transactions_df_imputed['product_id'] = Transactions_df_imputed['product_id'].astype(int)
Transactions_df_imputed['customer_id'] = Transactions_df_imputed['customer_id'].astype(int)
Transactions_df_imputed['online_order'] = Transactions_df_imputed['online_order'].astype(int)
```
